```python
# read filter coefficients from a .fcf file generated by Matlab

import numpy as np
import matplotlib.pyplot as plt

b = np.loadtxt('filter-fir-coeff1.fcf', skiprows=14)

print("Filter length: {}".format(len(b)))
```

```python
%matplotlib inline

plt.stem(b)
```

```python
from scipy.signal import freqz

w, h = freqz(b)

plt.plot(w, 20*np.log(np.abs(h)))
```
